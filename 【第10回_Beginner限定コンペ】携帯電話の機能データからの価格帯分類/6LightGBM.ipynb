```python
import pandas as pd
train = pd.read_csv("train.csv", index_col=0) # 学習用データ
test = pd.read_csv("test.csv", index_col=0)   # 評価用データ
sample_submit = pd.read_csv("sample_submission.csv", index_col=0, header=None) # 応募用サンプルファイル
```

```python
train.head()
```

```python
y = train["price_range"]
X = train.drop(["price_range"],axis=1)
```

```python
import lightgbm as lgb #LightGBM
# モデルの学習
model = lgb.LGBMClassifier() # モデルのインスタンスの作成
model.fit(X, y) # モデルの学習
# テストデータの予測クラス (予測クラス(0 or 1 or 2 or 3)を返す)
pred = model.predict(test)
```
